```python
# Get file list to pull from
file_list = []
with os.scandir(path) as it:
    x = map(lambda x: x.name, it)
    file_list = list(x)
```

```python
data_dict = {}
game_list = []
batch_list = []
```

```python
def json_parser(i, data_dict=data_dict):
    data_row = {'id': data_dict[i]['id'],
                'name': data_dict[i]['slug'],
                'released': data_dict[i]['released'],
                'rating': data_dict[i]['rating'],
                'ratings_count': data_dict[i]['ratings_count']
                           }
    return data_row
```

```python
for entry in file_list:
        with open('final\\data\\' + entry) as file:
            data_dict = pd.read_json(file, typ='records')
            x = map(
                lambda x: json_parser(x, data_dict=data_dict), range(len(data_dict))
            )
            game_df = game_df.append(list(x), ignore_index=True)
        print("Now processing: " + str(entry))
```

```python
game_df.to_parquet('game_data.parquet')
```
