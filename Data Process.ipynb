```python
import os
import numpy as np
import pandas as pd
import json
```

```python
column_names = ['id', 'name', 'released', 'rating', 'ratings_count']
```

```python
game_df = pd.DataFrame(columns=column_names)
```

```python


with open('final/data/data1.json') as file:
    data_dict = pd.read_json(file, typ='records')
    for i in range(len(data_dict)):
        
        data_row = { 'id': data_dict[i]['id'],
        'name': data_dict[i]['slug'],
        'released': data_dict[i]['released'],
        'rating': data_dict[i]['rating'],
        'ratings_count': data_dict[i]['ratings_count']
                   }
        game_df = game_df.append(data_row, ignore_index=True)
```
