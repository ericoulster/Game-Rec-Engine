```python
import numpy as np
import pandas as pd
```

```python
genre_df = pd.read_parquet('genre_list.parquet')
```

```python
genre_tab = pd.crosstab(genre_df['game_name'], genre_df['genre_name'])
```

```python
genre_tab
```

```python
from scipy.spatial.distance import pdist, squareform
```

```python
jaccard_distances = pdist(genre_tab.values, metric='jaccard')
```

```python
jaccard_similarity = 1 - jaccard_distances
```

```python
jaccard_square = squareform(jaccard_similarity)
```

```python
rec_table_v1 = pd.DataFrame(jaccard_square, index=genre_tab.index, columns=genre_tab.index)
```

```python
genre_tab.to_parquet('genre_crosstab.parquet')
```
